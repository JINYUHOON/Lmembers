```python
def ratio(df):
  test = df[['고객번호','대분류명']]
  test = test.set_index('고객번호')
  A = pd.get_dummies(test['대분류명']).reset_index()
  A = A.groupby('고객번호').sum().astype(int)
  A['total'] = A.sum(axis=1).astype(int)
  for i in Li:
    A[f"{i}_r"] = A[i] / A['total']
  B = A.drop(['가공식품', '가구/인테리어', '교육/문화용품', '기타', '디지털/가전', '명품', '신선식품', '의류',
    '일상용품', '전문스포츠/레저', '패션잡화','total'],axis=1)
  B['total'] = B.sum(axis=1)
  return B
```

```python
def pur_ratio(df):
  test = pd.get_dummies(df['대분류명'])
  d_test = df[['고객번호','구매금액','영수증번호']]
  ttest = pd.concat([d_test,test],axis=1)
  ttest= ttest[['고객번호', '영수증번호', '가공식품', '가구/인테리어', '교육/문화용품', '기타', '디지털/가전','명품', '신선식품', '의류', '일상용품', '전문스포츠/레저', '패션잡화', '구매금액']]
  tttest = ttest.groupby('고객번호').sum()
  a = pd.DataFrame()

  for i in Li: 
    R = ttest.groupby(['고객번호',i])[['구매금액']].sum()
    R = R.reset_index()
    R = R.rename(columns={'구매금액':f'{i}_구매금액'})
    R.loc[R[i]==0,f'{i}_구매금액'] = 0
    AA = R.groupby('고객번호')[[i,f'{i}_구매금액']].sum()
    a = pd.concat([a,AA[f'{i}_구매금액']],axis=1)

  Result = pd.concat([a,tttest],axis=1)
  Result = Result.drop('영수증번호',axis=1)
  Result = Result.rename(columns={'구매금액':'총구매금액'})
  Result = Result[['가공식품',
  '가구/인테리어',
  '교육/문화용품',
  '기타',
  '디지털/가전',
  '명품',
  '신선식품',
  '의류',
  '일상용품',
  '전문스포츠/레저',
  '패션잡화',
  '가공식품_구매금액',
  '가구/인테리어_구매금액',
  '교육/문화용품_구매금액',
  '기타_구매금액',
  '디지털/가전_구매금액',
  '명품_구매금액',
  '신선식품_구매금액',
  '의류_구매금액',
  '일상용품_구매금액',
  '전문스포츠/레저_구매금액',
  '패션잡화_구매금액',
  '총구매금액'
  ]]
  return Result
```
