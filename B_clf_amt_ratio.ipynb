```python
def setidx(df):
  df = df.reset_index()
  df = df.rename(columns={'index':'고객번호'})
  df = df.set_index('고객번호')
  return df

df_1 = setidx(pr_1)
df_2 = setidx(pr_2)
df_3 = setidx(pr_3)
df_4 = setidx(pr_4)
```

```python
sorting_cols = pr_1.columns.to_list()
sorting_cols = ['가공식품_구매금액',
 '가구/인테리어_구매금액',
 '교육/문화용품_구매금액',
 '기타_구매금액',
 '디지털/가전_구매금액',
 '명품_구매금액',
 '신선식품_구매금액',
 '의류_구매금액',
 '일상용품_구매금액',
 '전문스포츠/레저_구매금액',
 '패션잡화_구매금액',
 '총구매금액']
```

```python
def_cols = ['가공식품_구매금액','가구/인테리어_구매금액',
 '교육/문화용품_구매금액',
 '기타_구매금액',
 '디지털/가전_구매금액',
 '명품_구매금액',
 '신선식품_구매금액',
 '의류_구매금액',
 '일상용품_구매금액',
 '전문스포츠/레저_구매금액',
 '패션잡화_구매금액']
```

```python
sum_cols = ['가공식품_구매금액_r', '가구/인테리어_구매금액_r', '교육/문화용품_구매금액_r',
       '기타_구매금액_r', '디지털/가전_구매금액_r', '명품_구매금액_r', '신선식품_구매금액_r', '의류_구매금액_r',
       '일상용품_구매금액_r', '전문스포츠/레저_구매금액_r', '패션잡화_구매금액_r']
```

```python
confirm_cols = ['가공식품_구매금액_r', '가구/인테리어_구매금액_r', '교육/문화용품_구매금액_r',
       '기타_구매금액_r', '디지털/가전_구매금액_r', '명품_구매금액_r', '신선식품_구매금액_r', '의류_구매금액_r',
       '일상용품_구매금액_r', '전문스포츠/레저_구매금액_r', '패션잡화_구매금액_r', 'total']
```

```python
p1 = df_1[sorting_cols]
p2 = df_2[sorting_cols]
p3 = df_3[sorting_cols]
p4 = df_4[sorting_cols]
```

```python
p1
```

```python
def ratio(df):
  global sorting_cols, def_cols, sum_cols, confirm_cols
  s_df = df[sorting_cols]
  for i in def_cols:
    s_df[f'{i}_r'] = s_df[i] / s_df['총구매금액']
  s_df['total'] = s_df[sum_cols].sum(axis=1)
  s_df = s_df[confirm_cols]
  return s_df
```

```python
H1_r = ratio(p1)
H2_r = ratio(p2)
H3_r = ratio(p3)
H4_r = ratio(p4)
```

```python
H1_r
```

```python
def reName(a,df):
  cols = list(df.columns)
  for i in cols:
    df[f'{a}_{i}'] = df[i]
  df = df.drop(cols,axis=1)
  return df
```

```python
H1 = reName('H1',H1_r)
H2 = reName('H2',H2_r)
H3 = reName('H3',H3_r)
H4 = reName('H4',H4_r)
```

```python
def train_df(a,b,c):
    A = pd.concat([a,b],axis=1)
    train = pd.concat([A,c],axis=1)
    return train

train = train_df(H1,H2,H3)
train
```

```python
trainlist = ['H1','H2','H3']
AA = ratio(p1)
c_li = list(AA.columns)

trainList = []
for col in c_li:
  for h in trainlist:
    trainList.append(f'{h}_{col}')

train = train[trainList]
```

```python
train = train.drop(['H1_total','H2_total','H3_total'],axis=1)
train
```

```python
def test_df(a,b,c,d):
    A = pd.concat([a,b],axis=1)
    train = pd.concat([A,c],axis=1)
    test = pd.concat([train,d],axis=1)
    return test

test = test_df(H1,H2,H3,H4)
```
